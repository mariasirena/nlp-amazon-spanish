Existen algunas stopwords observadas que podrían ser de interés para el análisis

```python
stop_words = STOP_WORDS
nlp.max_length = 2000000

stop_words.remove("bueno")
stop_words.remove("demasiado")
stop_words.remove("buena")
stop_words.remove("buen")
stop_words.remove("buenos")
stop_words.remove("despacio")
stop_words.remove("pocos")
stop_words.remove("poca")
stop_words.remove("bien")
stop_words.remove("raras")
stop_words.remove("ninguno")
stop_words.remove("ningunas")
stop_words.remove("no")
stop_words.remove("si")
stop_words.remove("sí")
stop_words.remove('diferentes')
stop_words.remove('diferente')
stop_words.remove('nuevo')
```

______________________________

```python
from unicodedata import normalize
```

```python
#Normalizacion, eliminacion de tildes y carácteres como números

rev_body = []

for i in range(data.shape[0]):
    review = data.iloc[i].review_body
    review = re.sub(
        r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", 
        normalize( "NFD", review), 0, re.I
    )
    review = normalize( 'NFC', review)
    
    # Vamos a reemplazar los caracteres que no sean letras por espacios
    review=re.sub("[^a-zA-Z]"," ",str(review))
    
    rev_body.append(review)
```

```python
data['Body'] = pd.Series(rev_body) 
```

```python
data.head()
```

```python
#Letra minúscula, tokenización y eliminación de stopwords

rev_body = []

for i in range(data.shape[0]):
    review = data.iloc[i].Body
    
    #minuscula
    review = review.lower()
    
    review= nltk.word_tokenize(review)
    
    review = [palabra for palabra in review if not palabra in stop_words]
    
    rev_body.append(review)
```

```python
data['Body_norm'] = pd.Series(rev_body) 
```

```python
data.head()
```

```python
#lemmatización

texts_out=[]
    
for i in range(data.shape[0]):
    review = data.iloc[i].Body_norm
    
    review=nlp(' '.join(review))
    
    allowed_postags = ['NOUN','ADJ','VERB','ADV']
    
    texts_out.append([token.lemma_ for token in review if token.pos_ in allowed_postags])
```

```python
data['Body_lemma'] = pd.Series(texts_out)
```

```python
data.head()
```

Adjetivos

```python
#buscar adjetivos

adjetivos = []

for i in range(data.shape[0]):
    review = data.iloc[i].Body_lemma
    review=nlp(' '.join(review))
    review = [token for token in review if token.pos_ == "ADJ"]
    adjetivos.append(review)
```

```python
data["body_ADJ"] = pd.Series(adjetivos)
```

# Machine Learning

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = matriz_revs.toarray()
y = stars_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42, stratify = y)
```

```python
X_train.shape, X_test.shape
```

## BENCHMARK - Naïve Bayes (MultinomialNB)

Como modelo benchmark selecciono Naïve Bayes debido a que es el modelo más simple y rápido de clasificación multimodal como es el caso con el que contamos.

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
nb.fit(X_train,y_train)

y_train_pred = nb.predict(X_train)
y_test_pred = nb.predict(X_test)

print("Accuracy train GaussianNB: ", accuracy_score(y_train, y_train_pred))
print("Accuracy test GaussianNB: ", accuracy_score(y_test, y_test_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
def confusion(y_test,y_test_pred,X):
    names=["1 estrella","2 estrellas", "3 estrellas", "4 estrellas", "5 estrellas"]
    cm=confusion_matrix(y_test,y_test_pred)
    f,ax=plt.subplots(figsize=(8,8))
    sns.heatmap(cm,annot=True,linewidth=.5,linecolor="r",fmt=".0f",ax=ax)
    plt.title(X, size = 25)
    plt.xlabel("y_pred")
    plt.ylabel("y_true")
    ax.set_xticklabels(names)
    ax.set_yticklabels(names)
    plt.show()

    return
```

```python
confusion(y_test, y_test_pred, 'GaussianNB')
```
